## Q1: Mathematical Properties of Statistics

### Formulas Review

Sample mean: $m(X) = \dfrac{1}{N} \sum_{i=1}^N x_i$

Sample covariance: $\text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(y_i - m(Y))$

Sample variance: $s^2 = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))^2$

### 1.1: Show that $m(a + bX) = a + b \times m(X)$

**Proof:**

$$
\begin{align}
m(a + bX) &= \dfrac{1}{N} \sum_{i=1}^N (a + bx_i) \\
&= \dfrac{1}{N} \sum_{i=1}^N a + \dfrac{1}{N} \sum_{i=1}^N bx_i \\
&= \dfrac{1}{N} \cdot N \cdot a + b \cdot \dfrac{1}{N} \sum_{i=1}^N x_i \\
&= a + b \cdot m(X)
\end{align}
$$

**Interpretation:** The mean of a linear transformation is the same linear transformation of the mean. Adding a constant shifts the mean by that constant, and multiplying by a constant scales the mean by that constant.

### 1.2: Show that $\text{cov}(X, a+bY) = b \times \text{cov}(X,Y)$

**Proof:**

First, note that $m(a + bY) = a + b \times m(Y)$ from part 1.

$$
\begin{align}
\text{cov}(X, a+bY) &= \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))((a + by_i) - m(a+bY)) \\
&= \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))((a + by_i) - (a + b \cdot m(Y))) \\
&= \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(by_i - b \cdot m(Y)) \\
&= \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X)) \cdot b(y_i - m(Y)) \\
&= b \cdot \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(y_i - m(Y)) \\
&= b \times \text{cov}(X,Y)
\end{align}
$$

**Interpretation:** Covariance is not affected by adding constants (the constant $a$ disappears), but it scales linearly with multiplication. This makes sense because covariance measures the joint variability of two variables.

### 1.3: Show that $\text{cov}(a+bX, a+bX) = b^2 \text{cov}(X,X)$ and that $\text{cov}(X,X) = s^2$

**Proof - Part 1:**

Using the result from 1.2 with $Y = a + bX$:

$$
\begin{align}
\text{cov}(a+bX, a+bX) &= b \times \text{cov}(X, a+bX) \\
&= b \times b \times \text{cov}(X, X) \quad \text{(applying 1.2 again)} \\
&= b^2 \text{cov}(X,X)
\end{align}
$$

**Proof - Part 2:**

$$
\begin{align}
\text{cov}(X,X) &= \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(x_i - m(X)) \\
&= \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))^2 \\
&= s^2
\end{align}
$$

**Interpretation:** The covariance of a variable with itself is its variance. When we scale a variable by $b$, its variance scales by $b^2$ (variance measures squared deviations, so scaling compounds). Adding constants doesn't affect variance since it doesn't change spread.

### 1.4: Non-decreasing transformations and the median

**Question:** If we apply a non-decreasing transformation $g()$ to a variable, is the transformation of the median equal to the median of the transformed variable?

**Answer:** **Yes**, for non-decreasing transformations, $g(\text{median}(X)) = \text{median}(g(X))$.

**Proof/Explanation:**

Let $m$ be the median of $X$. By definition, at least 50% of values are $\le m$ and at least 50% are $\ge m$.

Since $g()$ is non-decreasing:
- If $x_i \le m$, then $g(x_i) \le g(m)$
- If $x_i \ge m$, then $g(x_i) \ge g(m)$

Therefore, at least 50% of the transformed values $g(x_i)$ are $\le g(m)$ and at least 50% are $\ge g(m)$, which means $g(m)$ is the median of $g(X)$.

**Does this apply to other quantiles?** **Yes!** The same logic applies to any quantile (e.g., 25th percentile, 75th percentile). Non-decreasing transformations preserve the ordering of data, so they preserve all quantiles.

**What about the IQR?** **Yes!** Since IQR = Q3 - Q1, and both Q3 and Q1 are preserved under non-decreasing transformations:
$$\text{IQR}(g(X)) = g(Q3) - g(Q1)$$

However, note that this is NOT the same as $g(\text{IQR}(X)) = g(Q3 - Q1)$ unless $g$ is also linear.

**What about the range?** **Yes!** Range = max - min. Since non-decreasing transformations preserve ordering:
$$\text{Range}(g(X)) = g(\max(X)) - g(\min(X))$$

Again, this is NOT the same as $g(\text{Range}(X))$ unless $g$ is linear.

### 1.5: Is $m(g(X)) = g(m(X))$ for non-decreasing transformations?

**Answer:** **No**, in general this is NOT true.

**Counterexample:**

Consider $X = \{1, 2, 3\}$ and $g(x) = x^2$.

- $m(X) = \frac{1+2+3}{3} = 2$
- $g(m(X)) = g(2) = 4$

But:
- $g(X) = \{1, 4, 9\}$
- $m(g(X)) = \frac{1+4+9}{3} = \frac{14}{3} \approx 4.67$

Clearly, $4 \neq 4.67$, so $g(m(X)) \neq m(g(X))$.

**When is it true?** This property ONLY holds for **linear** transformations of the form $g(x) = a + bx$. We proved in part 1.1 that for linear transformations, $m(a + bX) = a + b \times m(X) = g(m(X))$.

**Why does it fail for non-linear transformations?** The mean is sensitive to the spacing and magnitude of values. Non-linear transformations (like $x^2$, $\log(x)$, $\sqrt{x}$) change the relative spacing of values, which affects where the mean falls. This is related to **Jensen's Inequality**: for a convex function $g$, $m(g(X)) \ge g(m(X))$, and for concave functions, $m(g(X)) \le g(m(X))$.

```python
# Let's verify our counterexample numerically
X = np.array([1, 2, 3])
g = lambda x: x**2  # non-decreasing transformation

mean_X = X.mean()
g_of_mean = g(mean_X)

g_X = g(X)
mean_of_g = g_X.mean()

print(f"X = {X}")
print(f"g(x) = x^2")
print(f"\nm(X) = {mean_X}")
print(f"g(m(X)) = {g_of_mean}")
print(f"\ng(X) = {g_X}")
print(f"m(g(X)) = {mean_of_g}")
print(f"\nAre they equal? {np.isclose(g_of_mean, mean_of_g)}")
print(f"Difference: {abs(g_of_mean - mean_of_g):.4f}")
```

### 2.2: Histogram and Description of Foreign Gift Amount

### 2.3: Gift Type Distribution

```python
# Value counts and proportions
gift_type_counts = gifts_df['Gift Type'].value_counts()
gift_type_props = gifts_df['Gift Type'].value_counts(normalize=True) * 100

print("Gift Type Counts:")
print(gift_type_counts)
print("\nGift Type Proportions (%):")
print(gift_type_props)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Bar chart
gift_type_counts.plot(kind='bar', ax=axes[0], color='steelblue', edgecolor='black')
axes[0].set_xlabel('Gift Type')
axes[0].set_ylabel('Count')
axes[0].set_title('Distribution of Gift Types')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True, alpha=0.3)

# Pie chart
axes[1].pie(gift_type_counts, labels=gift_type_counts.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Proportion of Gift Types')

plt.tight_layout()
plt.show()

print("\nKEY FINDINGS:")
for gift_type, prop in gift_type_props.items():
    print(f"- {gift_type}: {prop:.2f}% of all gifts")
```

### 2.4: Kernel Density Plots of Log Gift Amount

```python
# Create log-transformed variable (add small constant to handle zeros)
gifts_df['log_gift_amount'] = np.log(gifts_df['Foreign Gift Amount'] + 1)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Overall KDE
gifts_df['log_gift_amount'].plot(kind='kde', ax=axes[0], linewidth=2, color='darkblue')
axes[0].set_xlabel('Log(Gift Amount)')
axes[0].set_ylabel('Density')
axes[0].set_title('Kernel Density Plot of Log Gift Amount')
axes[0].grid(True, alpha=0.3)

# KDE conditional on gift type
for gift_type in gifts_df['Gift Type'].unique():
    subset = gifts_df[gifts_df['Gift Type'] == gift_type]['log_gift_amount']
    subset.plot(kind='kde', ax=axes[1], linewidth=2, label=gift_type)

axes[1].set_xlabel('Log(Gift Amount)')
axes[1].set_ylabel('Density')
axes[1].set_title('Kernel Density Plot of Log Gift Amount by Gift Type')
axes[1].legend(title='Gift Type')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nPATTERNS OBSERVED:")
print("\nMedian log gift amount by type:")
print(gifts_df.groupby('Gift Type')['log_gift_amount'].median().sort_values(ascending=False))
print("\nMedian actual gift amount by type:")
print(gifts_df.groupby('Gift Type')['Foreign Gift Amount'].median().sort_values(ascending=False))
```

**Patterns observed:**
- The log transformation makes the distribution much more symmetric and normal-looking
- Different gift types show distinct distribution patterns
- Contracts tend to have different amount distributions compared to monetary gifts
- Real estate gifts (if present) typically represent larger amounts

### 3.3: Cross Tabulation of Room Type and Property Type

```python
# Cross tabulation
cross_tab = pd.crosstab(airbnb_df['Property Type'], airbnb_df['Room Type'], margins=True)
print("Cross Tabulation: Property Type x Room Type")
print(cross_tab)

# Proportions within each property type
cross_tab_prop = pd.crosstab(airbnb_df['Property Type'], airbnb_df['Room Type'], normalize='index') * 100
print("\n\nProportions (% within each Property Type):")
print(cross_tab_prop.round(2))
```

```python
# Visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Stacked bar chart - counts
cross_tab_for_plot = pd.crosstab(airbnb_df['Property Type'], airbnb_df['Room Type'])
cross_tab_for_plot.plot(kind='bar', stacked=True, ax=axes[0], colormap='Set2', edgecolor='black')
axes[0].set_xlabel('Property Type')
axes[0].set_ylabel('Count')
axes[0].set_title('Distribution of Room Types by Property Type (Stacked Counts)')
axes[0].legend(title='Room Type')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True, alpha=0.3, axis='y')

# Stacked bar chart - proportions
cross_tab_prop.plot(kind='bar', stacked=True, ax=axes[1], colormap='Set2', edgecolor='black')
axes[1].set_xlabel('Property Type')
axes[1].set_ylabel('Percentage')
axes[1].set_title('Distribution of Room Types by Property Type (Proportions)')
axes[1].legend(title='Room Type')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nKEY PATTERNS:")
print("\nProperty types where Private Room is more common than Entire home/apt:")
for prop_type in cross_tab_prop.index:
    if 'Private room' in cross_tab_prop.columns and 'Entire home/apt' in cross_tab_prop.columns:
        private_pct = cross_tab_prop.loc[prop_type, 'Private room']
        entire_pct = cross_tab_prop.loc[prop_type, 'Entire home/apt']
        if private_pct > entire_pct:
            print(f"- {prop_type}: {private_pct:.1f}% private room vs {entire_pct:.1f}% entire home/apt")
```

**Mean vs Median:**
The **median** appears to be a more reliable estimate of central tendency because:
1. The data are heavily right-skewed with extreme outliers
2. The mean is pulled upward by expensive listings (e.g., castles, boats)
3. The median is resistant to outliers and better represents the "typical" listing
4. Looking at the statistics, we consistently see mean > median, indicating right skew
5. The high standard deviations (often larger than the means) show extreme variability

For example, if a category has a mean of $200 but a median of $120, the median of $120 better represents what a typical user would pay.

### 4.5: Melt and Plot Onshore vs Offshore Rigs

```python
# Clean the onshore and offshore columns
onshore_col = 'Crude Oil and Natural Gas Rotary Rigs in Operation, Onshore (Number of Rigs)'
offshore_col = 'Crude Oil and Natural Gas Rotary Rigs in Operation, Offshore (Number of Rigs)'

rigs_df['Onshore'] = pd.to_numeric(rigs_df[onshore_col], errors='coerce')
rigs_df['Offshore'] = pd.to_numeric(rigs_df[offshore_col], errors='coerce')

print("Onshore Rigs Statistics:")
print(rigs_df['Onshore'].describe())
print("\nOffshore Rigs Statistics:")
print(rigs_df['Offshore'].describe())
```

```python
# Melt the data for plotting
melted_rigs = rigs_df[['time', 'Onshore', 'Offshore']].melt(id_vars='time', 
                                                              var_name='Location', 
                                                              value_name='Rig_Count')

print("Melted data structure:")
print(melted_rigs.head(10))
print(f"\nShape: {melted_rigs.shape}")
```

```python
# Plot the melted data
plt.figure(figsize=(15, 6))

# Plot both series
onshore_data = melted_rigs[melted_rigs['Location'] == 'Onshore']
offshore_data = melted_rigs[melted_rigs['Location'] == 'Offshore']

plt.plot(onshore_data['time'], onshore_data['Rig_Count'], 
         linewidth=2, label='Onshore', color='saddlebrown')
plt.plot(offshore_data['time'], offshore_data['Rig_Count'], 
         linewidth=2, label='Offshore', color='steelblue')

plt.xlabel('Year')
plt.ylabel('Number of Rigs')
plt.title('Onshore vs Offshore Rotary Rigs in Operation Over Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nCOMPARISON: Onshore vs Offshore")
print("\n1. SCALE:")
print(f"   - Onshore rigs are much more numerous (max ~{rigs_df['Onshore'].max():.0f})")
print(f"   - Offshore rigs are fewer (max ~{rigs_df['Offshore'].max():.0f})")
print(f"   - Onshore represents {(rigs_df['Onshore'].mean() / (rigs_df['Onshore'].mean() + rigs_df['Offshore'].mean()) * 100):.1f}% of total on average")

print("\n2. PATTERNS:")
print("   - Both show similar boom-bust cycle in 1970s-1980s")
print("   - Onshore declined more dramatically after the 1980s peak")
print("   - Offshore remained relatively more stable")

print("\n3. RECENT TRENDS:")
recent_onshore = rigs_df['Onshore'].iloc[-12:].mean()
recent_offshore = rigs_df['Offshore'].iloc[-12:].mean()
print(f"   - Recent onshore average (last year): {recent_onshore:.0f}")
print(f"   - Recent offshore average (last year): {recent_offshore:.0f}")
```

```python
# Create a stacked area plot for better visualization of composition
plt.figure(figsize=(15, 6))
plt.fill_between(rigs_df['time'], 0, rigs_df['Onshore'], 
                 alpha=0.7, label='Onshore', color='saddlebrown')
plt.fill_between(rigs_df['time'], rigs_df['Onshore'], 
                 rigs_df['Onshore'] + rigs_df['Offshore'], 
                 alpha=0.7, label='Offshore', color='steelblue')

plt.xlabel('Year')
plt.ylabel('Number of Rigs')
plt.title('Composition of Rotary Rigs: Onshore vs Offshore (Stacked)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
